```python
from pyspark.sql import functions as F

# raw filepaths
raw_file_customers = "/Volumes/dev/trailsport/raw_data/Customers.csv"
raw_file_products = "/Volumes/dev/trailsport/raw_data/Products.csv"
raw_file_sales = "/Volumes/dev/trailsport/raw_data/Sales.csv"

#bronze table filepaths
bronze_table_customers = "dev.trailsport.bronze_customers"
bronze_table_products = "dev.trailsport.bronze_products"
bronze_table_sales = "dev.trailsport.bronze_sales"
```

# 1. Load the data and write to bronze

```python
#load customer data, 
df = spark.read.csv(raw_file_customers, header=True, inferSchema=True).withColumn("ingest_time",F.current_timestamp())
#insert in bronze table
df.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable(bronze_table_customers)
```

```python
#load products data, 
df = spark.read.csv(raw_file_products, header=True, inferSchema=True).withColumn("ingest_time",F.current_timestamp())
#insert in bronze table
df.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable(bronze_table_products)
```

```python
#load sales data, 
df = spark.read.csv(raw_file_sales, header=True, inferSchema=True).withColumn("ingest_time",F.current_timestamp())
#insert in bronze table
df.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable(bronze_table_sales)
```
